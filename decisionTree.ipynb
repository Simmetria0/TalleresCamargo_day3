```python
#como siempre por primera cosa importamos la libreria
from sklearn import tree
import matplotlib.pylab as plt
import numpy as np
```

# CLASSIFICACION.



```python
#Creamos un datasample muy simple
X = [[0, 0], [1, 1]]
Y = [0, 1]

plt.plot(X[0][0], X[0][1], 'bo')
plt.plot(X[1][0], X[1][1], 'r+')


```

```python
plt.plot(X[0][0], X[0][1], 'bo')
plt.plot(X[1][0], X[1][1], 'r+')
plt.plot(2, 2)
plt.text(2, 2, u"?", fontsize=20)
```

```python
#Entrena la decision tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
#y ya puedes preguntarle que opina acerca de. la clase de (2,2)
clf.predict([[2., 2.]])
```

```python
#probamos con un dataset un poco mas grande
#creamos el dataset
size_sample = 100
mu1 = -2
sigma1 = 2
x1_1 = mu1 + sigma1*np.random.randn(size_sample)
x1_2 = mu1 + sigma1*np.random.randn(size_sample)
y1 = np.ones(size_sample)
mu2 = 6
sigma2 = 5
x2_1 = mu2 + sigma2*np.random.randn(size_sample)
x2_x  = mu2 + sigma2*np.random.randn(size_sample)
y2 = np.zeros(size_sample)
X = np.column_stack((np.append(x1_1,x2_1),np.append(x1_2,x2_x)))
y = np.append(y1,y2)
```

```python
#que pinta tiene?
import matplotlib.pyplot as plt
colors = ['red', 'blue']
#plt.scatter(X[:,0], X[:,1], c='g')
plt.scatter(X[y==0,0], X[y==0,1], c=colors[0])
plt.scatter(X[y==1,0], X[y==1,1], c=colors[1])

plt.show()
```

```python
#como antes, vamos a entrenar el decition tree:
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
#y vemos como predice puntos nuevos
clf.predict([[10., 10.]])
```

```python
#you can visualize the tree
tree.plot_tree(clf.fit(X, y)) 
```
