```python
#como siempre por primera cosa importamos la libreria
from sklearn import tree
import matplotlib.pylab as plt
import numpy as np
```

```python
X = [[0, 0], [2, 2]]
y = [0.5, 2.5]

```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X, y)
clf.predict([[1, 1]])
```
